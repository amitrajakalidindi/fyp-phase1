```python
import pandas as pd
import numpy as np
songs = pd.read_csv("songs.csv")
```

```python
songs.drop_duplicates(subset='title', keep="first", inplace=True)
songs.rename(columns = {'pop':'popularity'}, inplace = True)
songs.head(10)
```

```python
userPlaylist = pd.read_csv("userPlaylist.csv")
songsList = np.array(songs["title"])
def isAvailable(song):
  if song in songsList:
    return 1
  else:
    return 0
userPlaylist['data_availability'] = userPlaylist.apply(lambda user: isAvailable(user['song']), axis = 1)
userPlaylist
```

```python
userPlaylist = userPlaylist[userPlaylist.data_availability == 1]
userPlaylist.shape
```

```python
userPlaylist.drop_duplicates(subset=['song', 'userId'], keep="first", inplace=True)
```

```python
usersList = np.unique(userPlaylist['userId'])
validUsers = []
for i in usersList:
  if(len(np.array(userPlaylist[userPlaylist.userId == i])) >= 20):
    validUsers.append(i)
len(validUsers)
```

```python
userPlaylist = userPlaylist[userPlaylist.userId.isin(validUsers)]
userPlaylist.shape
```
